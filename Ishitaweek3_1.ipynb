```python
import numpy as np
import pandas as pd

data = pd.read_csv("/content/Data-cleaning-for-beginners-using-pandas.csv")
```

```python
data.columns = data.columns.str.lower().str.replace(' ', '_')
```

```python
data.head()
```

```python
data.isna().sum()
```

AGE

```python
avg = data.age.mean()
data.age = data.age.fillna(avg).round(decimals=1)
data.age = data.age.astype(int)
```

SALARY

```python
data.salary = data.salary.str.replace('k', '000')
data.salary = data.salary.str.replace('$', ' ')
data.salary
```

RATING

```python
avg = data.rating.mean()
data.rating = data.rating.replace(-1, np.nan)
data.rating = data.rating.fillna(avg).round(decimals=1)
data.rating
```

LOCATION

```python
data['location'] = data['location'].str.replace('Australia Aus' , 'Australia, Aus')
data.location = data.location.str.replace("India In", "India, In")
data.location
```

ESTABLISHED

```python
data.established = data.established.replace(-1, np.nan)
data.established.fillna('Unknown', inplace = True)
data.established
```

EASY APPLY

```python
data.easy_apply = data.easy_apply.str.replace('-1', 'TRUE')
data.easy_apply
```
